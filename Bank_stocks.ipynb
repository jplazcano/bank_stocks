```python
import pandas_datareader as pdr
import datetime as dt
import pandas
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
start = dt.datetime(2012,11,16)
end = dt.datetime(2022,11,16)
```

```python
jpm = pdr.DataReader('JPM', 'yahoo', start, end) # JP Morgan Chase historical data (10 years)
bac = pdr.DataReader('BAC', 'yahoo', start, end) # Bank of America historical data (10 years)
c = pdr.DataReader('C', 'yahoo', start, end) # CityGroup historical data (10 years)
wfc = pdr.DataReader('WFC', 'yahoo', start, end) # Wells Fargo historical data (10 years)
```

```python
jpm['Open'].plot(label = 'JP Morgan', figsize = (15,7))
bac['Open'].plot(label = 'Bank of America')
c['Open'].plot(label = 'Citygroup')
wfc['Open'].plot(label = 'Wells Fargo')
plt.title('Stock Prices of JPM, BAC ,IDCBY and WFC')
plt.legend()
```

With this preliminary price summary, we can look into how the stocks changed over time. 

The open stock prices of all four companies since 16th November of 2012 are displayed in the graph above.
When compared to the other three, JP Morgan has had the highest open pricing while Bank of America the lowest.

While Bank of America had been pretty consistent over the 10 years, Jp Morgan, Citygroup  and Wells Fargo experienced biggest fluctuations.

```python
jpm['Volume'].plot(label = 'JP Morgan', figsize = (16,8))
bac['Volume'].plot(label = "Bank of America")
c['Volume'].plot(label = 'Citygroup')
wfc['Volume'].plot(label = 'Wells Fargo')
plt.title('Volume of Stock Traded')
plt.legend()
```

The graph up top compares the volume of trading for the four firms over the previous 10 years.
The most actively traded stock is Bank of America.

```python
#Market Capitalization
jpm['Market_Cap'] = jpm['Open'] * jpm['Volume']
bac['Market_Cap'] = bac['Open'] * bac['Volume']
c['Market_Cap'] = c['Open'] * c['Volume']
wfc['Market_Cap'] = wfc['Open'] * wfc['Volume']
jpm['Market_Cap'].plot(label = 'JP Morgan', figsize = (16,8))
bac['Market_Cap'].plot(label = 'Bank of America')
c['Market_Cap'].plot(label = 'Citygroup')
wfc['Market_Cap'].plot(label = 'Wells Fargo')
plt.title('Market Capitalization')
plt.legend()
```

Volume or stock prices are not always an effective approach to compare different companies.
To determine a firm's value, one can look at its market capitalization, which is the sum of the value of its shares on a publicly traded firm. 

To more effectively compare the companies, I have plotted the market cap of Volume * Share price in the graph above. JP Morgan appears to be valued higher than the other three firms while Wells Fargo the less valued.

```python
#Calculating returns to measure volatility

jpm['returns'] = np.log(jpm['Close']).diff()
bac['returns'] = np.log(bac['Close']).diff()
c['returns'] = np.log(c['Close']).diff()
wfc['returns'] = np.log(wfc['Close']).diff()

jpm['returns'].hist(bins = 50, label = 'JP Morgan', figsize = (16,8))
bac['returns'].hist(bins = 50, label = 'Bank of America')
c['returns'].hist(bins = 50, label = 'Citygroup')
wfc['returns'].hist(bins = 50, label = 'Wells Fargo')
plt.legend()
```

The volatility measures the rate of fluctuations in the stock prices over time and indicates the level of risk associated with the price changes of the stock.
Calculating the percentage rise in stock value from the previous day can be used to measure volatility.
The stock price of Citygroup appearsthe most volatile among the four companies compared.

```python
print('Statistics of JP Morgan daily returns:')
print(jpm['returns'].describe()*100)
print("")
print('Statistics of Bank of America daily returns:')
print(bac['returns'].describe()*100)
print("")
print('Statistics of Citygroup daily returns:')
print(c['returns'].describe()*100)
print("")
print('Statistics of Wells Fargo daily returns:')
print(wfc['returns'].describe()*100)
```
